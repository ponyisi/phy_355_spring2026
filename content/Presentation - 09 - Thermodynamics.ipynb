"Thermodynamics": the observed macroscopic behavior of materials, heat, and energy exchange. (If you're trying to design steam engines in the Industrial Revolution, this is the level of description that is important.)

"Heat", from a microscopic perspective: energy held in externally hard-to-see motions
* in thermodynamics energy is separated into "heat" (energy transferred between objects by thermal contact), "internal energy", and "work" (some macroscopic motion, such as an expanding gas pushing a piston).
* so-called "internal energy" &mdash; was accepted somewhat slowly, in parallel with the picture of materials being made of atoms

We can relate "big picture" observable properties to the microscopic description coming from statistical mechanics.

Let's look at the "Laws of Thermodynamics."  (One thing about these laws: they have many, many alternative statements...)

## The idea of Work

"Work": all kinds of energy transfer that are not heat.
* work can be used to lift a weight, turn a motor, etc. - it is "useful" energy
* reflects energy that is organized in some way. 

For example: a pressurized gas in a chamber might push a piston outwards 
* can then be used to lift a weight)
* the kinetic energy of the random motions of the molecules of the gas is turned into the organized coherent motion of the atoms of the piston

If the center of mass of the piston is moving, the number of internal states available to the atoms of the piston does not change, so neither does its entropy. (In other words, the temperature of an object doesn't depend if it's moving.) So the energy associated with the bulk motion of the piston is work, as opposed to the random motion of the atoms relative to the piston's center of mass motion, which are heat.

Work isn't necessarily mechanical; it can be converted into other kinds of energy (for example generating electric potential energy in an electric generator).

## Heat Engines

Critical question in thermodynamics: the conversion of heat energy into work. 
* clearly possible: a steam locomotive boils water, which expands as it turns into gas, and the pressure differential with the outside air can be used to make pistons move (creating work). 

Are there general laws that govern how to design such systems, and how we can make them maximally efficient? 
* regardless of design, there are fundamental limits on how well we can convert heat into work.

In fact this is still a critical question today; many electric power technologies (fossil fuel, nuclear, geothermal, solar thermal) work by making something hot and turning that heat into work.  The conceptual thermodynamic limitations that applied to the earliest steam engines still apply to such systems.

## Entropy and Heat Transfer

We defined temperature via

$$ \frac{1}{T} = \frac{\partial S}{\partial E} $$

If we deal in very small amounts of internal energy change $\Delta E$ and entropy change $\Delta S$ in a system, the definition of temperature implies

$$ 
\begin{align*}
\frac{1}{T} &= \frac{\Delta S}{\Delta E}\\
\Rightarrow \Delta S &= \frac{\Delta E}{T}
\end{align*}
$$

So if I increase the internal energy of a system by an amount $\Delta E$, I also increase the entropy $S$ of the system by an amount $\Delta E/T$. The higher the temperature of the system, the smaller the entropy change for the same energy change.

## The Zeroth Law
### "All thermometers are the same", "All heat is of the same kind"

If systems A and B are in thermal equilibrium, and that B and C are in thermal equilibrium, then A and C are in thermal equilibrium.  

Our definition of thermal equilibrium here is that their temperatures are the same.  
* in statistical mechanics the definition of temperature comes from the relationship between the number of states and the internal energy, so this is automatic
* not obvious before people had a modern understanding of heat as atomic motion

## The First Law
### "Energy is conserved"
This is the statement that if a system has internal energy $U$, I put heat $Q$ into it, and the system does work $W$ on its surroundings, then

$$ Q = \Delta U + W $$



## The Second Law
###  "Entropy increases", "Heat moves from hotter to colder systems"
* the total entropy of an isolated system does not decrease over time
* the total entropy of an isolated system will strictly increase unless a process is of a special kind called "reversible." 

In practice essentially everything interesting is not a reversible process - in particular no real heat engine is reversible.

We have already seen that the definition of temperature is such that moving energy from a hotter system to a colder one will increase the combined entropy of the two; this satisfies the Second Law.

> “The law that entropy always increases holds, I think, the supreme position among the laws of Nature. If someone points out to you that your pet theory of the universe is in disagreement with Maxwell's equations - then so much the worse for Maxwell's equations. If it is found to be contradicted by observation - well, these experimentalists do bungle things sometimes. But if your theory is found to be against the Second Law of Thermodynamics I can give you no hope; there is nothing for it to collapse in deepest humiliation.” &mdash; Arthur Eddington

## The Third Law
### "You can't reach absolute zero"

This one is a bit more abstract than the others, but basically it says that no physical process can cool a system to absolute zero, because the amount of energy you can remove from a system decreases as the temperature decreases.

## Heat Engines and the Laws of Thermodynamics

A heat engine: a device that takes heat from a thermal reservoir at some temperature and produces work.
* we will see shortly that a heat engine will also need a heat reservoir to dump residual heat into
* the engine is assumed to return to its original state after going through some number of internal states (its *cycle*). 
  * there are many cycles that have been devised, often lending their name to specific kinds of engine - the Diesel cycle, the Stirling cycle, the Carnot cycle, the Otto cycle (gas engines)...

![Engine schematic](images/engine_schematic.png)

### The First Law

We can't get more work out of a heat engine than we put heat energy in; we would violate conservation of energy otherwise.  Considering a realistic engine (where there's a hot reservoir, a cold reservoir, and work being done), the heat energy removed from the hot reservoir ($Q$) must be equal to the sum of the heat energy added to the cold reservoir ($Q'$) and the work done ($W$).

Informally, as far as heat engines are concerned, the First Law is summarized as "you can't win"; your engine can't make more energy in work than it takes in as heat.

### The Second Law

The fact that the total entropy increases strongly affects how heat engines can operate.  
* the work that an engine produces has no entropy associated with it (in our piston example, the energy that goes into getting the piston to move doesn't contribute to the entropy of the piston)
* if I could just extract heat from a reservoir and turn it 100% into work with an engine that went back to its original state after a cycle, then the entropy of the heat reservoir would decrease (removing energy from it, $\Delta E$ and hence $\Delta S$ would be negative) but the engine would not change entropy &mdash; the total entropy would decrease. This is forbidden.

This is the reason why heat engines need to dump energy into a cold reservoir. This _increases_ the entropy of the cold reservoir. Let's suppose the reservoirs are at temperatures $T_H$ and $T_C$ for hot and cold, respectively. Removing energy $Q$ from the hot reservoir reduces its entropy by $Q/T_H$. If we try to generate exactly that much entropy by putting energy into the cold reservoir, we need to only put in the energy $Q'_\mathrm{min}$ defined by

$$ 
\begin{align*}
\frac{Q'_\mathrm{min}}{T_C} &= \Delta S_\mathrm{cold} = -\Delta S_\mathrm{hot} = \frac{Q}{T_H} \\
Q'_\mathrm{min} &= \frac{T_C}{T_H} Q
\end{align*}
$$

If we have $T_C < T_H$ then $Q'_\mathrm{min} < Q$, and the difference $Q-Q'_\mathrm{min}$ can be extracted and turned into work.

This thermodynamic argument tells us that for any kind of heat engine we have a limit on the possible efficiency, i.e. what fraction of input heat energy $Q$ can be turned into useful work $W$:

$$ \eta_\mathrm{max} = \frac{W_\mathrm{max}}{Q} = \frac{Q-Q'_\mathrm{min}}{Q} = 1 - \frac{T_C}{T_H} $$

This is called the *Carnot efficiency*, and no heat engine can do better.  For a thermal power plant with $T_H \sim$ 850 K and $T_C \sim$ 300 K, this gives a maximum possible efficiency $\sim$ 64%. (In practice 35-45% is more normal.) You do better with a hotter hot reservoir or a colder cold reservoir.

The informal statement of the Second Law is "you can't break even" &mdash; converting heat completely to work would reduce total entropy, and to avoid this we need to lose fraction of the heat to a reservoir.

### The Third Law

"But wait!", one might say. The equation above says $\eta_\mathrm{max} = 1$ if $T_C = 0$. (If a system is at absolute zero, then adding any energy increases the entropy by infinity, via $\Delta S = \Delta E/T$, so I could dump arbitrarily small amounts of energy into the cold reservoir and keep the total entropy increasing.)  What about this option?

This is foreclosed by the Third Law. If you can't actually make a system at temperature absolute zero, then it is irrelevant whether you could make a heat engine that operated with a cold reservoir at $T_C = 0$.  This is why the informal statement of the Third Law is "you can't get out of the game" &mdash; the loophole of constructing a 100%-efficient engine with a cold reservoir at absolute zero is impossible.

## The (Thermodynamic) Arrow of Time

Newton's laws do not distinguish past and future, and neither do any of the other laws of classical physics. If it were technically possible to reverse the direction of motion of every single particle and electromagnetic wave in the universe, we would see time go "backward".

For example, a block sliding slowly down an incline plane converts its gravitational potential energy to heat energy in the plane (the heat is generated by friction). 
* It is in principle possible for all the molecules in the plane to reverse direction and transfer energy to the block, pushing it up the ramp! But this would be a phenomenally implausible occurrence, because the molecules would have to be in an extremely special configuration to make this happen.  
* So we expect that blocks sliding down ramps will be slowed and stopped by friction (warming the ramp), not that the ramp will get colder and start pushing the block up, even though theoretically the latter is possible. 
* The fundamental reason is that converting the gravitational potential energy to heat _increases_ entropy, while spontaneously converting heat to gravitational potential energy _decreases_ it, so the Second Law permits the first but not the second.

Therefore in a very real sense the difference between "past" and "future" is just the difference between "lower entropy" and "higher entropy." The thermodynamic arrow of time is the fact that the future is different from the past not from some underlying asymmetry in the microscopic laws, but in the number of accessible states.

One thing that isn't addressed by this is what sets the low entropy state of the past in the first place; the implication is that the Big Bang was a state of very low entropy. Why this should be so is an active research question.

## Heat capacity

For many systems the energy $\Delta E$ required to heat them up by $\Delta T$ is approximately a constant, and the two are related by the formula

$$ \Delta E = C \Delta T $$

where $C$ is called the _heat capacity_.  For example, it takes 1 calorie = 4.184 J of energy to raise the temperature of one gram of water by one degree Kelvin, so the heat capacity of liquid water is 1 cal/K = 4.184 J/K.  (Note this has the same units as Boltzmann's constant...)

Heat capacities are measured experimentally, but statistical mechanics gives us a way to relate it to underlying microscopic physics. In ideal gases, for example, assuming volume and number of molecules are held constant, the Sackur-Tetrode equation (and the equipartition principle) gave us

$$ E = \frac{3}{2} N k_B T = \frac{3}{2} n R T$$

where $E$ was the total kinetic energy of the molecules of the ideal gas and $N$ was the number of molecules.  This implies that 

$$ C = \frac{3}{2} N k_B = \frac{3}{2} n R $$

for an ideal gas; expressing this as a _molar_ heat capacity (i.e. the heat capacity per mole, in units J K<sup>-1</sup> mol<sup>-1</sup>), we would get 

$$C_\mathrm{molar} = \frac{3}{2} R = 12.5\ \textrm{J K}^{-1}\textrm{ mol}^{-1}$$

This is correct for the noble gases at STP!  These are single atoms, so our state counting assumption is correct.  More complex molecules, such as nitrogen (N<sub>2</sub>) have _higher_ heat capacities &mdash; for diatomic gases this changes with temperature and tends to saturate at $\frac{7}{2} R$ &mdash; which tells us that something more complex is going on (the simple ideal gas model is undercounting the places energy can go).

https://en.wikipedia.org/wiki/Molar_heat_capacity#/media/File:DiatomicSpecHeat2.png

For crystalline solids, a similar kind of rule applies, except that the molar heat capacity $\approx 3R$.  This is called the *Dulong-Petit Law*. The extra factor of two comes from the fact that there is energy associated with the displacement of the atoms from their equilibrium positions in the crystalline lattice - so there are six degrees of freedom per atom (the momentum of the atom and the position of the atom), giving us $6 \times \frac{1}{2} k_B T$ average energy per atom.

```python
# Here we will generate plots of the heat capacity of various elements
# data are from NIST, documented on data.gov:
# https://catalog.data.gov/dataset/the-nbs-tables-of-chemical-thermodynamic-properties-selected-values-for-inorganic-and-c1-a
import pandas
if not JUPYLITE:
    df = pandas.read_csv('https://data.nist.gov/od/ds/mds2-2124/NBS_Tables_with_ErratumValues.csv',
                        header=1, index_col='Formula')
else:
    df = pandas.read_csv('https://ponyisi.github.io/phy_355_spring2026/lab/data/NBS_Tables_with_ErratumValues.csv',
                        header=1, index_col='Formula')

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8,6)
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['animation.embed_limit'] = 60
```

```python
R=8.314 # J/mol K
# Select noble gases in the gaseus state, force the columns we want to be a float
noble_gases = df[df['State']=='g'].loc[['He', 'Ne', 'Ar', 'Kr', 'Xe']].astype({'Cp\nJ mol-1 K-1': 'float', 'Molar Mass\ng mol-1': 'float'})
#print(noble_gases)
# The following is heat capacity at *constant pressure* - which is the *constant volume* heat capacity + one
noble_gases['Cp/R']=noble_gases['Cp\nJ mol-1 K-1']/R
# Select some other gases, force the columns we want to be a float
other_gases = df[df['State']=='g'].loc[['O2', 'CO', 'Cl2', 'H2S', 'CO2', 'CH4', 'CCl4']].astype({'Cp\nJ mol-1 K-1': 'float', 'Molar Mass\ng mol-1': 'float'})
other_gases['Cp/R']=other_gases['Cp\nJ mol-1 K-1']/R
fig, ax = plt.subplots(figsize=(15,6))
other_gases.plot(x='Molar Mass\ng mol-1', y='Cp/R', kind='scatter', color='r', ax=ax)
noble_gases.plot(x='Molar Mass\ng mol-1', y='Cp/R', kind='scatter', color='b', ax=ax)#, figsize=(8,6))
for row in noble_gases.itertuples():
    ax.annotate(row.Index, (row._5, row._13), xytext=(5, 0), textcoords='offset points')
for row in other_gases.itertuples():
    ax.annotate(row.Index, (row._5, row._13), xytext=(5, 0), textcoords='offset points')
plt.ylabel('$C_p/R = C_V/R + 1$', fontsize='x-large')
plt.xlabel('Molar weight', fontsize='x-large')
plt.axhline(2.5)
ax.set_ylim(2,11)
plt.show()
```

```python
# Select atomic solids
solids = df[df['State']=='cr']
solid_elements=solids[(solids['ΔfH°']=='0') & (solids['Cp\nJ mol-1 K-1']!='-')]
solid_monatomic=solid_elements[solid_elements.index.map(lambda x: x.isalpha())].astype({'Cp\nJ mol-1 K-1': 'float', 'Molar Mass\ng mol-1': 'float'})
#solids[solids.index.map(lambda x: len(x)<=2) & (solids)]
solid_monatomic['Cp/R']=solid_monatomic['Cp\nJ mol-1 K-1']/R
plot=solid_monatomic.plot(x='Molar Mass\ng mol-1', y='Cp/R', kind='scatter', figsize=(15,6))
exceptions=solid_monatomic[(solid_monatomic['Cp/R']<2.5) | (solid_monatomic['Cp/R']>3.5)]
for row in exceptions.itertuples():
    plot.annotate(row.Index, (row._5, row._13), xytext=(5, 0), textcoords='offset points', fontsize='large')
plt.ylabel('$C_p/R$', fontsize='x-large')
plt.xlabel('Atomic weight', fontsize='x-large')
plt.axhline(3)
plt.show()
```
